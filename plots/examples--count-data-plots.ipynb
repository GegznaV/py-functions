## Example

```python
# Packages and modules --------------------------------
import pandas as pd
import matplotlib.pyplot as plt

import plot_counts as my

# Settings --------------------------------------------
# Default plot options
plt.rc("figure", titleweight="bold")
plt.rc("axes", labelweight="bold", titleweight="bold")
plt.rc("font", weight="normal", size=10)
plt.rc("figure", figsize=(7, 3))
```

[Palmer penguins](https://allisonhorst.github.io/palmerpenguins/articles/intro.html) dataset will be used to demonstrate the functions.

```python
# Data
url = "https://vincentarelbundock.github.io/Rdatasets/csv/palmerpenguins/penguins.csv"
penguins = pd.read_csv(url, index_col=0)
```

```python
penguins.sample(5)
```

### Plot Single Variable

```python

counts_island = my.calc_counts_and_percentages("island", data =  penguins)
my.plot_counts_with_labels(
    counts_island, ylabel="Count", title="Island of Penguin's Residence", rot=0
)

display(counts_island)
```

```python
counts_sex = my.calc_counts_and_percentages("sex", data = penguins)
my.plot_counts_with_labels(
    counts_sex, ylabel="Count", title="Penguin's Sex", rot=0, color="limegreen"
)

display(counts_sex)
```

### Plot Variable Pairs

```python
my.plot_crosstab_as_barplot(
    x="species", 
    y="island", 
    data=penguins,
    title="Penguin Species by Island"
);
```

```python
my.plot_crosstab_as_barplot(
    x="species",
    y="sex",
    data=penguins,
    normalize="all",
    title="Penguin Species by Sex"
);
```

```python
my.plot_crosstab_as_barplot(x="year", y="island", data=penguins, ylim=[None, 90]);
```
